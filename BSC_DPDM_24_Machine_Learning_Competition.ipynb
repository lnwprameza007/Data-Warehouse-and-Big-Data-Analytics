```python
import pandas as pd

train = pd.read_csv('/content/train_set.csv')
test = pd.read_csv('/content/X_test.csv')
```

```python
train
```

```python
train.info()
```

```python
# ตรวจสอบค่าสถิติพื้นฐานของ Xn
print(train["Xn"].describe())

# ตรวจสอบความสัมพันธ์ระหว่าง Xn กับตัวแปรอื่น ๆ
print(train.corr(numeric_only=True)["Xn"].sort_values(ascending=False))

# ตรวจสอบจำนวนค่าที่ไม่ซ้ำกันใน Xn
print(f"Unique values in Xn: {train['Xn'].nunique()}")
```

```python
for col in ["Xn", "Hb(Gm%)", "Hct(%)", "MCV(fl)", "RBC", "MCH", "MCHC"]:
    train[col].fillna(train[col].median(), inplace=True)
```

```python
for col in ["A2 (2-3%)", "HbF (2%)", "HbA (95-97%)"]:
    train[col] = train.groupby("Thal_Group")[col].transform(lambda x: x.fillna(x.mean()))
```

```python
train.info()
```

```python
train = train.drop(columns=['MCV-RBC-3*HGB', 'MCV-RBC-5*HGB', '|80-MCV|', '|27-MCH|', '|80-MCV|*|27-MCH|'], errors='ignore')
```

```python
train.info()
```

```python
train = train.drop(columns= ['id number'])
```

```python
train.columns
```

```python
train['Thal_Group'].unique()
```

```python
# prompt: เปลี่ยน คอลัมภ์ Thal_Group เป็นตัวเลข โดย 'Abnormal_variants' = 0 'Beta_thal_related' = 1 'HbE_related' = 3

train['Thal_Group'] = train['Thal_Group'].replace({'Abnormal_variants': 0, 'Beta_thal_related': 1, 'HbE_related': 3, 'Alpha_thal_related':4, 'Complex_genotypes':5, 'Normal':6})
```

```python
train
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10, shuffle=True, random_state=6)
```

```python
X = train.drop(columns=['Thal_Group'])
y = train['Thal_Group'].astype(int)
```

```python
#import
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#defind
dt_classifier = DecisionTreeClassifier() #default

#train #evaluation
cv_scores = cross_val_score(dt_classifier, X, y, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(dt_classifier, X, y, cv=kf, scoring='f1_macro')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())

print("F1 scores:", f1_scores)
print("Average F1 score:", f1_scores.mean())
print("Standard deviation of F1 scores:", f1_scores.std())
```

```python
#import
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#defind
st_classifier = DecisionTreeClassifier(max_depth=6) #default

#train #evaluation
cv_scores = cross_val_score(st_classifier, X, y, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(st_classifier, X, y, cv=kf, scoring='f1_macro')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())

print("F1 scores:", f1_scores)
print("Average F1 score:", f1_scores.mean())
print("Standard deviation of F1 scores:", f1_scores.std())
```

```python
#import
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#defind
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42) #default

#train #evaluation
cv_scores = cross_val_score(rf_classifier, X, y, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(rf_classifier, X, y, cv=kf, scoring='f1_macro')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())

print("F1 scores:", f1_scores)
print("Average F1 score:", f1_scores.mean())
print("Standard deviation of F1 scores:", f1_scores.std())
```

```python
#import
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#defind
nb_classifier = GaussianNB()

#train #evaluation
cv_scores = cross_val_score(nb_classifier, X, y, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(nb_classifier, X, y, cv=kf, scoring='f1_macro')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())

print("F1 scores:", f1_scores)
print("Average F1 score:", f1_scores.mean())
```

```python
#import
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#defind
svm_classifier = SVC(kernel='rbf', C=1.0, gamma='scale')

#train #evaluation
cv_scores = cross_val_score(svm_classifier, X, y, cv=kf, scoring='accuracy')
f1_scores = cross_val_score(svm_classifier, X, y, cv=kf, scoring='f1_macro')

#Print Cross-validation scores
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())

print("F1 scores:", f1_scores)
print("Average F1 score:", f1_scores.mean())
```
